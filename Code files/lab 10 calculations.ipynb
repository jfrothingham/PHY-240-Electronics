```python
# import statements
import matplotlib.pyplot as plt
import numpy as np
```

```python
# collected data
data = np.loadtxt('vd_vr_data.csv', skiprows=1, delimiter=',')
vr = data[:,2]   # volts
vd = data[:,3]   # volts
```

```python
# calculate current --> vr = i * R
r = 330   # Ohms
i = vr / r
```

```python
plt.title('Diode I-V characteristic from collected data', fontsize=15)
plt.xlabel('$V_D$ [volts]', fontsize=12)
plt.ylabel('current I [amp]')
plt.plot(vd,i,'-')
```
