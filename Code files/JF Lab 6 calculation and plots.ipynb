```python
# import statements
import matplotlib.pyplot as plt
import numpy as np
```

```python
# collected peak-to-peak voltage data from low frequency to high frequency
freq = np.array([50, 100, 500, 1000, 10000, 50000]) #Hz
Vr = np.array([0.47,0.92,4.19,6.58,9.06,8.62]) #v
Vc = np.array([9.45,9.34,8.55,6.85,1.05,0.24]) #v
Vs = np.array([9.59,9.55,9.65,9.63,9.17,8.65]) #v
```

```python
# calculating gain
gain_vr = Vr / Vs
gain_vc = Vc / Vs
```

```python
# plotting gain vs frequency
plt.figure(figsize=(15,7))
plt.suptitle('Gain vs Frequency',fontsize=20)

plt.subplot(1,2,1)
plt.title('x-axis log scale',fontsize=15)
plt.ylabel('gain')
plt.xlabel('frequency [Hz]')
plt.semilogx(freq, gain_vr, 'r*--', label='$V_r$')
plt.semilogx(freq, gain_vc, 'b*--', label='$V_c$')
plt.legend()
plt.grid()

plt.subplot(1,2,2)
plt.title('both axes log scale',fontsize=15)
plt.ylabel('gain')
plt.xlabel('frequency [Hz]')
plt.loglog(freq, gain_vr, 'r*--', label='$V_r$')
plt.loglog(freq, gain_vc, 'b*--', label='$V_c$')
plt.legend()
plt.grid()
```

```python
# collected time difference data from low frequency to high frequency
tdiff_vr = np.array([0.004138, 0.003285, 0.000347, 0.000123, 0.00000174, 0]) #s
tdiff_vc = np.array([-0.001074, -0.000284, -0.000126, -0.000126, -0.000023, -0.000004]) #s
```

```python
# calculating phase difference
phase_diff_vr = freq * tdiff_vr
phase_diff_vc = freq * tdiff_vc
```

```python
# plotting phase difference vs frequency
plt.title('Phase difference vs Frequency',fontsize=20)
plt.ylabel('phase difference [radians]')
plt.xlabel('frequency [Hz]')
plt.semilogx(freq, phase_diff_vr, 'r*--', label='$V_r$')
plt.semilogx(freq, phase_diff_vc, 'b*--', label='$V_c$')
plt.legend()
plt.grid()
```
