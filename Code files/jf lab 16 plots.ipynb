```python
# import statements
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# COLLECTED DATA

# no filter
no_data = np.loadtxt('gain_phase_nofilter.csv', skiprows=1, delimiter=',')
no_freq = no_data[:,1]
no_mag = no_data[:,2]
no_phase = no_data[:,3]

# low-pass filter
lo_data = np.loadtxt('gain_phase_lowpass.csv', skiprows=1, delimiter=',')
lo_freq = lo_data[:,1]
lo_mag = lo_data[:,2]
lo_phase = lo_data[:,3]

# high-pass filter
hi_data = np.loadtxt('gain_phase_highpass.csv', skiprows=1, delimiter=',')
hi_freq = hi_data[:,1]
hi_mag = hi_data[:,2]
hi_phase = hi_data[:,3]

# both filters
both_data = np.loadtxt('gain_phase_bothfilters.csv', skiprows=1, delimiter=',')
both_freq = both_data[:,1]
both_mag = both_data[:,2]
both_phase = both_data[:,3]
```

```python
# GAIN AND PHASE VS FREQUENCY PLOTS

flat = np.average(no_mag)
critical = flat - 3

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.title('Gain vs frequency', fontsize=15)
plt.ylabel('Gain magnitude in dB', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(no_freq, no_mag, 'b-', label='No filters')
plt.semilogx(lo_freq, lo_mag, 'g-', label='Low-pass filter')
plt.semilogx(hi_freq, hi_mag, 'r-', label='High-pass filter')
plt.semilogx(both_freq, both_mag, 'm-', label='Both filters')
plt.axvline(54)
plt.axvline(2500)
plt.axhline(critical)
plt.legend(fontsize=12)
plt.grid()

plt.subplot(1,2,2)
plt.title('Phase difference vs frequency', fontsize=15)
plt.ylabel('Phase difference in degrees', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(no_freq, no_phase, 'b-', label='No filters')
plt.semilogx(lo_freq, lo_phase, 'g-', label='Low-pass filter')
plt.semilogx(hi_freq, hi_phase, 'r-', label='High-pass filter')
plt.semilogx(both_freq, both_phase, 'm-', label='Both filters')
plt.legend(fontsize=12)
plt.grid()
```

```python
plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.title('Gain vs frequency (No filters)', fontsize=15)
plt.ylabel('Gain magnitude in dB', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(no_freq, no_mag, 'b-', label='No filters')
plt.ylim([-15,7])
plt.grid()

plt.subplot(2,2,2)
plt.title('Gain vs frequency (Low-pass filter)', fontsize=15)
plt.ylabel('Gain magnitude in dB', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(lo_freq, lo_mag, 'g-', label='Low-pass filter')
plt.ylim([-15,7])
plt.grid()

plt.subplot(2,2,3)
plt.title('Gain vs frequency (High-pass filter)', fontsize=15)
plt.ylabel('Gain magnitude in dB', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(hi_freq, hi_mag, 'r-', label='High-pass filter')
plt.ylim([-15,7])
plt.grid()

plt.subplot(2,2,4)
plt.title('Gain vs frequency (Both filters)', fontsize=15)
plt.ylabel('Gain magnitude in dB', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(both_freq, both_mag, 'm-', label='Both filters')
plt.ylim([-15,7])
plt.grid()
```

```python
plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.title('Phase difference vs frequency (No filters)', fontsize=15)
plt.ylabel('Phase difference in degrees', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(no_freq, no_phase, 'b-', label='No filters')
plt.ylim([-180,180])
plt.grid()

plt.subplot(2,2,2)
plt.title('Phase difference vs frequency (Low-pass filter)', fontsize=15)
plt.ylabel('Phase difference in degrees', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(lo_freq, lo_phase, 'g-', label='Low-pass filter')
plt.ylim([-180,180])
plt.grid()

plt.subplot(2,2,3)
plt.title('Phase difference vs frequency (High-pass filter)', fontsize=15)
plt.ylabel('Phase difference in degrees', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(hi_freq, hi_phase, 'r-', label='High-pass filter')
plt.ylim([-180,180])
plt.grid()

plt.subplot(2,2,4)
plt.title('Phase difference vs frequency (Both filters)', fontsize=15)
plt.ylabel('Phase difference in degrees', fontsize=12)
plt.xlabel('Frequency in Hz', fontsize=12)
plt.semilogx(both_freq, both_phase, 'm-', label='Both filters')
plt.ylim([-180,180])
plt.grid()
```
